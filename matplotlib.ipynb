```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
xpoints = np.array([1,8])
ypoints= np.array([3,10])

plt.plot (xpoints,ypoints,"o")
plt.show()
```

```python
xpoints = np.array([1,8])
# ypoints= np.array([3,10])

plt.plot (xpoints,marker= "o")
plt.show()
```

```python
xpoints = np.array([1,8,4,5,6])
# ypoints= np.array([3,10])

plt.plot (xpoints,marker= "o")
plt.show()
```

```python
xpoints = np.array([1,8])
# ypoints= np.array([3,10])

plt.plot (xpoints,marker= "*")
plt.show()
```

```python
xpoints = np.array([1,8,3,4,5,6])
# ypoints= np.array([3,10])

plt.plot (xpoints,"o:b")
plt.show()
```

```python
xpoints = np.array([1,8,3,4,5,6])
# ypoints= np.array([3,10])

plt.plot (xpoints,marker = "*",ms = 30)
plt.show()
```

```python
xpoints = np.array([1,8,3,4,5,6])
# ypoints= np.array([3,10])

plt.plot (xpoints,marker = "*",ms = 30,mec="g")
plt.show()
```

```python
xpoints = np.array([1,8,3,4,5,6])
# ypoints= np.array([3,10])

plt.plot (xpoints,marker = "*",ms = 30,mec="g",mfc ="r")
plt.show()
```

```python
xpoints = np.array([1,8,3,4,5,6])
# ypoints= np.array([3,10])

plt.plot (xpoints, ls ="--")
plt.show()
```

```python
# xpoints = np.array([1,8,3,4,5,6])
ypoints= np.array([3,10,4,5])

plt.plot (ypoints, linewidth ="15")
plt.show()
```

```python
xpoints = np.array([1,8,3,4,5,6])
ypoints= np.array([3,10,4,5])

plt.plot (ypoints)
plt.plot(xpoints)
plt.xlabel("XPOINTS")
plt.ylabel("ypoints")
plt.title("lineplot")
plt.grid()    
plt.show()
```
